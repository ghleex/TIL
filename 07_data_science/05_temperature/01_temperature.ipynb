* 일교차가 가장 큰 시기는 언제?
* 겨울에는 어느 시기가 가장 추울까?

```python
import csv

f = open('wj_temperature.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

max_temp = -999
max_date = ''
for row in data:
    if row[-1] == '':
        row[-1] = -999
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
#     print(row)
f.close()
print(max_date, max_temp)
```

```python
f = open('wj_temperature.csv', encoding='cp949')
data = csv.reader(f)
next(data)

results = [float(row[-1]) for row in data if row[-1] if row[0].split('-')[1] == '08']
# print(results)
# print(len(results))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 5))
plt.plot(results, 'hotpink')
plt.show()
```
