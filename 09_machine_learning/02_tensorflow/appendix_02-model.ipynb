## 1. 보스턴 집값 예측

```python
# 라이브러리
import tensorflow as tf
import pandas as pd
```

```python
# 1. 데이터
경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
보스턴 = pd.read_csv(경로)
```

```python
# 독립변수, 종속변수
독립 = 보스턴[[
    'crim', 'zn', 'indus', 'chas', 'nox', 
    'rm', 'age', 'dis', 'rad', 'tax',
    'ptratio', 'b', 'lstat'
]]
종속 = 보스턴[['medv']]
print(독립.shape, 종속.shape)
```

```python
# 2.1. 모델 구조 만들기 - 기존 방식
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 3.1. 데이터를 이용하여 모델 학습(FIT)
model.fit(독립, 종속, epochs=1000)
```

```python
model.predict(독립[:5])
```

```python
# 2.2. `BatchNormalization` layer 를 이용하여 모델 구조 만들기
X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 3.2. 데이터를 이용하여 모델 학습(FIT)
model.fit(독립, 종속, epochs=1000)
```

```python
model.predict(독립[:5])
```

## 2. 아이리스 품종 분류

```python
# 라이브러리
import tensorflow as tf
import pandas as pd
```

```python
# 1. 데이터
경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스 = pd.read_csv(경로)
```

```python
# 원핫인코딩
아이리스 = pd.get_dummies(아이리스)
```

```python
# 독립변수, 종속변수
독립 = 아이리스[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 아이리스[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(독립.shape, 종속.shape)
```

```python
# 2.1. 모델 구조 만들기 - 기존 방식
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(
    loss='categorical_crossentropy',
    metrics='accuracy'
)
```

```python
# 3.1. 데이터를 이용하여 모델 학습(FIT)
model.fit(독립, 종속, epochs=1000)
```

```python
model.predict(독립[:5])
```

```python
# 2.2. `BatchNormalization` layer 를 이용하여 모델 구조 만들기
X = tf.keras.layers.Input(shape=[4])
 
H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)
 
H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)
 
H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)
 
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(
    loss='categorical_crossentropy',
    metrics='accuracy'
)
```

```python
# 3.2. 데이터를 이용하여 모델 학습(FIT)
model.fit(독립, 종속, epochs=1000)
```

```python
model.predict(독립[:5])
```
