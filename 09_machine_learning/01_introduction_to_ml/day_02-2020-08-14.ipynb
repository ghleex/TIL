# 2. 애플리케이션? 프로그램?

* 애플리케이션 - 프로그램: 같은 것을 가리키는 다른 표현
  * 애플리케이션: 어떤 기능을 부품으로 응용하여 만든 완제품
    * 머신러닝의 "모델"이라는 부품을 응용하여 만든 소프트웨어 - 머신러닝 애플리케이션
    * 부품을 잘 활용하면, 원리를 알지 못해도 능력을 활용할 수 있음
  * 프로그램: 시간의 순서에 따라 동작하는 것
    * 과거에는 음악회에서 시간의 순서에 따라 곡이 연주되는 것을 일컬었음

* 프로그램은 아래의 조건 속에서도 일을 효율적으로 처리할 수 있음
  * 수백만 개의 작은 작업을 순서대로
  * 작업 누락되면 안 됨
  * 순서가 달라지면 안 됨
  * 하루에 수십 번씩 반복되며, 매일 해야 함
  * 모든 사람이 해야 함
  * 작업 종료를 예측하기 어려움
  * 실수가 일어나면 안 됨
  * 실수하기 쉬움
* 기계가 해야 할 일을 기계가 알아들을 수 있는 방식으로 순서대로 적는 것이 프로그램!

# 3. 마법인가? 기술인가?

## 사물인터넷
* 사물인터넷(Internet of Things): 인터넷으로 사물을 제어
* 전등, 자동차, 도로교통시스템, 국가 행정 시스템 등 모두 프로그램으로 제어할 수 있음
* 사물에 컴퓨터가 들어가면 손쉽게 기능을 제어할 수 있음
* 컴퓨터가 인터넷에 연결되어 있다면 지구 반대편에서도 해당 장치를 제어할 수 있음. 즉, 원격으로 조종할 수 있는 앱을 만들 수 있다는 것

## 사물인터넷 + 머신러닝?
* 머신러닝 도입 시
  * 어떤 기능을 실행할 것인지 스스로 판단할 수 있게 됨
  * 장치에 연결된 여러 센서를 통해 데이터를 수집,
  * 데이터의 의미를 머신러닝의 모델이 판단
  * 그 판단의 결과에 따라 여러 결정을 기계 스스로 할 수 있게 됨

* 사물의 원격 조종 + 지능 부여
* 사물인터넷 = 코딩 + 네트워크 + 전자공학 + 기계공학
* 우리는 데이터만으로 할 수 있는 일을 찾아보자

## 레모네이드 판매를 위한 레몬의 양 예측해보기
* 필요한 레몬의 양 예측 어떻게?
* 온도와 판매량 간 관계? 온도에 따라 판매량 변화?
* 실제로는 다양한 요소와 관계되어 예측하기 매우 까다로움
  * 복잡한 현실을 반영한 데이터를 이용하여 사람이 모델을 만드는 일은 어려움
* 머신러닝을 이용한다면 머신러닝이 자동으로 모델을 만들어줄 수 있음

### Example
* **레모네이드 카페**
  * 환경: 레모네이드 카게를 운영하고 있습니다. 처음에는 손님이 있다는 것만으로도 행복했습니다.
  * 불만족
    * 사업이 커지면서 레모네이드 판매량을 예측하기 어려워졌습니다. 감에 의존해서 레몬을 구입했지만 번번이 틀립니다. 레몬을 적게 사온 날엔 손님을 돌려보내야 했습니다. 레몬을 많이 사온 날엔 레몬을 버려야 했습니다. 수요를 예측하지 못해 생기는 손실이 눈덩이처럼 커지기 시작했습니다.
    
  * 꿈: 일기예보의 온도에 따라 그날 몇 개의 레몬이 필요할지 보다 정확히 예측하고 싶습니다.
